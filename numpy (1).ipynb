NumPy Universal Functions (ufuncs) are functions that operate element-wise on arrays. They provide a fast and efficient way to perform mathematical operations on arrays.

```python
import numpy as np
a = np.array([1,2,3,4,5,6])
b=np.array([7,8,9,10,11,12])
h=[]
l=[]
for i in range(len(a)):
  h.append(a[i]+b[i])
  l.append(a[i]*b[i])
print('addition:',h)
print('multiplication:',l)
```

```python
import numpy as np
a = np.array([1,2,3,4,5,6])
b=np.array([7,8,9,10,11,12])
w=[]
e=[]
t=[]
y=[]
for i,j in zip(a,b):
  n=i+j
  w.append(n)
  r=i*j
  e.append(r)
  s=i-j
  t.append(s)
  u=i/j
  y.append(u)
print('addition:',np.array(w))
print('multiplication:',np.array(e))
print('substraction:',np.array(t))
print('division:',np.array(y))
```

```python
import numpy as np
a = np.array([1,2,3,4,5,6])
b=np.array([7,8,9,10,11,12])
print('addition:',np.add(a,b),'multiplication:',np.multiply(a,b),'substraction:',np.subtract(a,b),'division:',np.divide(a,b),'power:',np.power(a,b),'remainder:',np.remainder(a,b),'mod:',np.mod(a,b),'quotient(divmod):',np.divmod(a,b))
```

```python
#union1d(This function finds the union of two arrays. It returns the unique,
# sorted array of values that are in either of the two input arrays.)
import numpy as np
a=np.array([[1,2,3,4,5],[11,12,13,14,15]])
b=np.array([6,7,8,5,10])
np.union1d(a,b)
```

```python
#intersection(intersect1d)(This function finds the intersection of two arrays.
#It returns the sorted, unique values that are in both of the input arrays.)
import numpy as np
a=np.array([1,2,3,4,5])
b=np.array([6,2,8,3,1])
np.intersect1d(a,b,assume_unique=True)
```

```python
#setdiff1d(This function finds the set difference of two arrays.
 #It returns the unique values in the first array that are not in the second array.)
import numpy as np
a=np.array([1,2,3,4,5])
b=np.array([6,2,8,3,1])
np.setdiff1d(a,b,assume_unique=True)
```

```python
#setxor1d(numpy.setxor1d() finds the exclusive or of two arrays,
#returning the sorted, unique values that are in only one (not both) of the input arrays.)
import numpy as np
a=np.array([1,2,3,4,5])
b=np.array([6,2,8,3,1])
np.setxor1d(a,b,assume_unique=True)
```
