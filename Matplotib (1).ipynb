Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide range of plotting functions to create various types of charts and graphs, including line plots, scatter plots, bar charts, histograms, and more.

*Key features and uses of Matplotlib:*

 - **Creating visualizations**: Matplotlib is primarily used for generating
visualizations of data to help understand trends, patterns, and relationships.

 - **Data exploration and analysis**: Visualizing data using Matplotlib aids in data exploration and analysis, making it easier to identify insights.

 - **Publication-quality figures**: Matplotlib allows customization of plots to create high-quality figures suitable for publications and presentations.

 - **Interactive plots**: Matplotlib supports interactive features, enabling users to zoom, pan, and interact with plots dynamically.

 - **Integration with other libraries**: Matplotlib integrates well with other scientific computing libraries in Python, such as NumPy and SciPy.

In essence, Matplotlib is a versatile and powerful tool for creating visualizations in Python, widely used in data science, scientific research, and various other fields.

#Size

You can change the size of the dots in your scatter plot using the s argument

#Matplotlib Bars

Matplotlib's bar() function creates vertical bar charts, while barh() creates horizontal ones, visualizing data with rectangular bars.

##Univariate



When you use a single set of values for the heights of the bars, the plot displays one variable (height) against different categories. It helps visualize the distribution of a single variable.

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.show()
```

##Stacked bar plot

**Bivariate**

In some cases, bar charts can show the relationship between two variables. This usually involves grouped or stacked bar charts. For example, you can use bars of different colors (or stacked on top of each other) to represent another dimension of data within each category

```python
x = np.array(["A", "B", "C", "D"])
y1 = np.array([3, 8, 1, 10])
y2 = np.array([2, 5, 3, 7])
plt.bar(x, y1, label='Group 1')
plt.bar(x, y2, bottom=y1, label='Group 2')
plt.legend()
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y1 = np.array([10, 20, 10, 30])
y2 = np.array([2, 5, 3, 7])
plt.bar(x, y1, label='Group 1',color='r')
plt.bar(x, y2, bottom=y1, label='Group 2',color='b')
plt.legend()
plt.show()
```

##Multiple Bar Plot

```python

# set width of bar
barWidth = 0.25
fig = plt. subplots(figsize = (12, 8))

# set height of bar
IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]

# Set position of bar on X axis
br1 = np. arange (len (IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, IT, color ='r', width = barWidth,
        edgecolor ='grey', label ='IT')
plt.bar(br2, ECE, color = 'g', width = barWidth,
        edgecolor = 'grey', label = 'ECE')
plt. bar(br3, CSE, color ='b', width = barWidth,
        edgecolor = 'grey', label = '(SE')
# Adding Xticks
plt. xlabel ('Branch', fontweight = 'bold', fontsize = 15)
plt.ylabel ('Students passed', fontweight = 'bold', fontsize = 15)
plt. xticks([r + barWidth for r in range (len (IT))],
            [' 2015', '2016', '2017', '2018', '2019'])
plt. legend ()
plt. show()
```

#Histogram

A histogram is a graphical representation of the distribution of numerical data. It groups data into bins (intervals) and uses bars to show the frequency of data points within each bin. The height of each bar represents the number of data points falling within that bin's range.

```python
x = np.random.normal(170, 10, 250)
plt.hist(x,edgecolor='black')
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data,bins=30,color = 'skyblue',edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = np.random.randn(1000)
sns.histplot(data,bins=30,kde = True,color = 'lightgreen',edgecolor='red')
plt.xlabel('Value')
plt.ylabel('Dencity')
plt.title('Costumaized Histogram with Dencity Plot')
plt.show()
```

```python
data1=np.random.randn(1000)
data2=np.random.normal(loc=3,scale=1,size=1000)

fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,4))

axes[0].hist(data1,bins=30,color='#0f6587',edgecolor='black')
axes[0].set_title('Histogram 1')

axes[1].hist(data2,bins=30,color='#acc94b',edgecolor='black')
axes[1].set_title('Histogram 2')

for ax in axes:
  ax.set_xlabel('Value')
  ax.set_ylabel('Frequency')


plt.tight_layout()
plt.show()
```

```python
data1=np.random.randn(1000)
data2=np.random.normal(loc=3,scale=1,size=1000)

plt.hist([data1,data2],bins=30,stacked=True,color=['cyan','Purple'],edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Stacked Histogram')
plt.legend(['Dataset 1','Dataset 2'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Generate sample data
x=np.linspace(0,2*np.pi,10)
y=np.sin(x)
# Create the stem plot
plt.stem(x,y)
# Add title and labels
plt.title('Stem Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate sample time series data
dates=pd.date_range(start='2024-01-01',periods=10,freq='D')
values=np.random.randn(10) # Random values for the time series

# Convert dates to numeric values (required for stem plot)
x=np.arange(len(dates))

# Create a stem plot
plt.stem(x,values,linefmt='g--',markerfmt='bo',basefmt='r--')

# customize the x-axis with the actual date labels
plt.xticks(x,[date.strftime('%Y-%m-%d') for date in dates],rotation=45)

# Add labels and tittle
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Stem Plot')
plt.tight_layout()

# Display the plot
plt.show()
```
